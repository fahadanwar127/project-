# Training Model Using Support Vector Classifier

```python
print("-- Training using SVC on Training Data --")
print("-- Parameters & Values: ", end='')

model_svc = SVC(gamma='auto', random_state=0)
model_svc.fit(train_x, np.ravel(train_y))

print(model_svc)
```

```python
# Saving Trained Model
pickle.dump(model_svc, open('model_svc.pkl', 'wb'))
```

```python
# Load saved Model
model = pickle.load(open('model_svc.pkl', 'rb'))
```

```python
model_predictions = model.predict(test_x)

testdata_predict = testdata.copy(deep=True)
pd.options.mode.chained_assignment = None

testdata_predict['Prediction'] = model_predictions
```

```python
# Printing Testing Data
print("-- Testing Data with Prediction --")
pd.set_option("display.max_rows", None, "display.max_columns", None)
testdata_predict
```

```python
model_accuracy_score = accuracy_score(testdata_predict['animal_name'], testdata_predict['Prediction'])

print("-- Model Accuracy Score: ", end='')
print(round(model_accuracy_score,3))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import joblib

# Example: Train and save a RandomForest model
iris = load_iris()
X, y = iris.data, iris.target
model = RandomForestClassifier()
model.fit(X, y)

# Save the model
joblib.dump(model, "model.pkl")
```

```python
import streamlit as st
import pandas as pd
import joblib

# Load the trained model
model = joblib.load("model.pkl")

# User Interface
st.title("Prediction Application")

st.sidebar.header("Input Features")
# Create inputs for the features
feature1 = st.sidebar.number_input("Feature 1 (e.g., Sepal Length)", value=5.1)
feature2 = st.sidebar.number_input("Feature 2 (e.g., Sepal Width)", value=3.5)
feature3 = st.sidebar.number_input("Feature 3 (e.g., Petal Length)", value=1.4)
feature4 = st.sidebar.number_input("Feature 4 (e.g., Petal Width)", value=0.2)

# Button to make predictions
if st.sidebar.button("Predict"):
    # Prepare the input data
    input_data = pd.DataFrame([[feature1, feature2, feature3, feature4]],
                              columns=["Feature 1", "Feature 2", "Feature 3", "Feature 4"])

    # Make prediction
    prediction = model.predict(input_data)
    prediction_proba = model.predict_proba(input_data)

    # Display the prediction
    st.write("### Prediction")
    st.write(f"Predicted Class: {prediction[0]}")
    st.write("### Prediction Probability")
    st.write(prediction_proba)

# Instructions
st.sidebar.info("Adjust the feature values to make predictions.")
```
